```python
############## Select rows where Text is not NaN

coders=coders.loc[coders["Text"].notna(),:]

##############  select all the rows except the last 3 that dont have index

coders=coders.iloc[:1344,:]
```

```python
############# drop unwanted columns

coders=coders.drop(['A','Seocondary Emotions', 'Emotion Primary', 'Emotion Secondary',
       'Secondary Emotion'], axis=1)
```

```python
################ i made a list of the id's but its a list of lists 

ID_list=[]
for i in range(len(filenames)):
    ID_list.append([filenames[i][1:8]]*96)
    
    
################# To flatten the list of lists i used a flatten method

def flatten(list_of_lists):
    if len(list_of_lists) == 0:
        return list_of_lists
    if isinstance(list_of_lists[0], list):
        return flatten(list_of_lists[0]) + flatten(list_of_lists[1:])
    return list_of_lists[:1] + flatten(list_of_lists[1:])


######### flatten with the method call and create the new column

ID_list=flatten(ID_list) 
coders['ID'] =ID_list    
```

```python
########################### Boolean array for iloc

a=np.array([True]*1344)
for i in range(1344):
    if(int(coders.ID[i][6])<5):  #########  if last digit <5 the assigned value is G2
        a[i]=True
    else:
        a[i]=False

        
```

```python
########################### initialize everything false and change the  <5 values to G2


Group_list=[['G1']*1344]
Group_list=flatten(Group_list)
len(Group_list)
coders.loc[:,['Group']]=Group_list

coders.iloc[a,5]='G2'


```

```python
########################### make it string to apply strip method and remove spaces before and after m + - and 0

coders.iloc[:,2]= coders.iloc[:,2].astype(str)
coders.iloc[:,2]=coders.iloc[:,2].str.strip()
```

```python
############################    replace Sentiment


coders.iloc[:,2]=coders.iloc[:,2].replace(["+","-", "m"], [1, -1, 1])
coders.iloc[:,2]=coders.iloc[:,2].replace(["+","_", " m"], [1, -1, 1]) ### there was an "_" value i counted it as "-"
coders.iloc[:,2]=coders.iloc[:,2].replace(["nan"], [np.nan])   ######### the nan values of sentiment became string so
#i reassign them as np.nan to compute the nan values in the next question
```

For question 5,i computed kappa for every possible annotator with each other (for every group and feeling).This way every kappa is computed twice but the mean isnt affected.Also i avoided computations of annotator's with themselves. 


```python
############################    5   

############### For group 1 and Sentiment 

Ids=["3662006","3662009","3662015","3662016","3662018"]       
a=[]
for IDD in Ids:
    for Id in Ids:
        if Id!=IDD:
            a.append(kappa(coders.loc[("G1",IDD),"Sentiment"].astype('int64').values,
                    coders.loc[("G1",Id),"Sentiment"].astype('int64').values))

np.mean(a)            
```

```python
############### For group 1 and Subjectivity

Ids=["3662006","3662009","3662015","3662016","3662018"]
a=[]
for IDD in Ids:
    for Id in Ids:
        if Id!=IDD:
            a.append(kappa(coders.loc[("G1",IDD),"Subjectivity"].astype('int64').values,
                    coders.loc[("G1",Id),"Subjectivity"].astype('int64').values))

np.mean(a)     
```

```python
############### For group 2 and Sentiment

Ids=["3662001","3662003","3662010","3662011","3662013","3662014","3662020","3662021","3662023"]
a=[]
for IDD in Ids:
    for Id in Ids:
        if Id!=IDD:
            a.append(kappa(coders.loc[("G2",IDD),"Sentiment"].astype('int64').values,
                    coders.loc[("G2",Id),"Sentiment"].astype('int64').values))

np.mean(a)     
```

```python
############### For group 2 and Subjectivity

Ids=["3662001","3662003","3662010","3662011","3662013","3662014","3662020","3662021","3662023"]
a=[]
for IDD in Ids:
    for Id in Ids:
        if Id!=IDD:
            a.append(kappa(coders.loc[("G2",IDD),"Subjectivity"].astype('int64').values,
                    coders.loc[("G2",Id),"Subjectivity"].astype('int64').values))

np.mean(a)     
```
